```python
# Permute list of numbers in Lexicographic order
# Worst Case Time Complexity is O(n!)
def LexPermutation(numList):
    
    length = (len(numList) - 1)
    p_of_x = -1
    i = 1
    
    # Look for highest index X such that P[X] < P[X + 1]
    while i <= length:
        if (numList[i - 1] < numList[i]):
            p_of_x = numList[i - 1]
            xIndex = i - 1
        i += 1
    
    # If there is no such X, P is the last permutation
    if p_of_x == -1: 
        return numList

    length = (len(numList))
    p_of_y = -1
    j = 1
    
    # Look for largest index Y such that P[X] < P[Y] 
    while j <= length:
        if (numList[j - 1] > p_of_x): 
            p_of_y = numList[j - 1]
            yIndex = j - 1
        j += 1

    # Swap X and Y
    numList[xIndex], numList[yIndex] = numList[yIndex], numList[xIndex]
    
    # Reverse P[X + 1...n]
    firstHalf = numList[:xIndex+1]
    secondHalf = numList[xIndex+1:]
    secondHalf.reverse()    
    newList = firstHalf + secondHalf
    print(newList)
    
    # Recursive call
    numList = LexPermutation(newList)
    return numList
```
